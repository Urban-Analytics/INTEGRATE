```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

tbo_vals = pd.read_csv('/nfs/a319/gy17m2a/PhD/datadir/RainGauge/interarrival_thresholds_CDD_noMissing.txt')

# Create figure and BNG-projected axis
fig = plt.figure(figsize=(7, 12))
ax = plt.axes(projection=ccrs.OSGB())  # EPSG:27700 (BNG)

# Add features (optional, since not all global ones show well in UK projection)
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle=':', edgecolor='gray')
ax.add_feature(cfeature.LAND, facecolor='white')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

# Plot WGS84 points (Cartopy transforms them automatically)
scatter = ax.scatter(
    tbo_vals['Lon'],
    tbo_vals['Lat'],
    color='black',
    s=15,
    marker='s',
    alpha=1,
    edgecolor=None,
    transform=ccrs.PlateCarree())

scatter = ax.scatter(
    tbo_vals['Lon'],
    tbo_vals['Lat'],
    c=tbo_vals['Critical_interarrival_time'],
    cmap='viridis_r',
    s=10,
    marker='.',
    alpha=1,
    edgecolor=None,
    transform=ccrs.PlateCarree()  # Tell Cartopy these are in WGS84
)

# Add tall vertical colorbar
axins = inset_axes(ax, width="3%", height="100%", loc='right',
                   bbox_to_anchor=(0.1, 0., 1, 1),
                   bbox_transform=ax.transAxes, borderpad=0)
cbar = plt.colorbar(scatter, cax=axins)
cbar.set_label("Critical Interarrival Time")

ax.text(
    0.01, 0.99,  # x, y in axes coordinates (near top-left)
    "(b)", 
    transform=ax.transAxes,
    fontsize=25, 
    fontweight='bold', 
    va='top', 
    ha='left',
    zorder=10  # ensure it's on top
)

# Title
# ax.set_title("Critical Interarrival Time (BNG Projection)")

plt.show()
fig.savefig("/nfs/a319/gy17m2a/PhD/Scripts/CatchmentAnalysis/PlotsForPaper/Figs/MIT_values.png", dpi=300, bbox_inches="tight",
           facecolor='white')
```

```python
from mpl_toolkits.axes_grid1 import AxesDivider, Size
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd

tbo_vals = pd.read_csv('/nfs/a319/gy17m2a/PhD/datadir/RainGauge/interarrival_thresholds_CDD_noMissing.txt')

fig = plt.figure(figsize=(7, 12))
ax = plt.axes(projection=ccrs.OSGB())

ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle=':', edgecolor='gray')
ax.add_feature(cfeature.LAND, facecolor='white')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

scatter = ax.scatter(
    tbo_vals['Lon'], tbo_vals['Lat'],
    c=tbo_vals['Critical_interarrival_time'],
    cmap='viridis_r',
    s=10,
    marker='.',
    alpha=1,
    edgecolor=None,
    transform=ccrs.PlateCarree()
)

# Create AxesDivider to add a colorbar axes beside the main axes
divider = AxesDivider(fig, ax, horizontal=False)

# Append a vertical axes for colorbar with fixed size
cax = divider.append_axes("right", size=Size.Fixed(30), pad=0.1)

# Create colorbar in the new axes
cbar = plt.colorbar(scatter, cax=cax)
cbar.set_label("Critical Interarrival Time")

ax.text(0.01, 0.99, "(b)", transform=ax.transAxes,
        fontsize=25, fontweight='bold', va='top', ha='left', zorder=10)

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx

# Load a world map and extract the UK
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
uk = world[world['name'] == 'United Kingdom']

# Reproject to Web Mercator (required for contextily basemaps)
uk_web_merc = uk.to_crs(epsg=3857)

# Plot
fig, ax = plt.subplots(figsize=(6, 5))
uk_web_merc.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=1.5)

# Add basemap
cx.add_basemap(ax, source=cx.providers.OpenTopoMap)

# Add label "(c)." in bold at top left
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
ax.text(xmin + 0.01*(xmax - xmin), ymax - 0.02*(ymax - ymin), 
        '(c).', fontsize=14, fontweight='bold', ha='left', va='top')

ax.set_axis_off()
plt.show()
fig.savefig("/nfs/a319/gy17m2a/PhD/Scripts/CatchmentAnalysis/PlotsForPaper/Figs/topo.png", dpi=300, bbox_inches="tight",
           facecolor='white')
```
